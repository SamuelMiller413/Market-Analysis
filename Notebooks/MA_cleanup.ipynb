# Script
Running this cell will turn the CSV into a Pandas dataframe, transform the data into binary values, and then save a new CSV File. 

```python
# import
import pandas as pd
# clone repository into Colab
!git clone https://github.com/SamuelMiller413/Market-Analysis
# I/O
    # from:
path_from = "/content/Market-Analysis/MA_Data.csv"
    # to:
path_to = "/content/Market-Analysis/MA_Data_Binary.csv"
# read CSV
MA_ = pd.read_csv(filepath_or_buffer=path_from)
# replace 'checked' with 1
MA_ = MA_.replace(to_replace="checked", value=1)
# impute missing data with 0
MA_ = MA_.fillna(0)
# drop 'Notes' column
MA_ = MA_.drop('Notes', axis=1)
# drop 'Z Z' row
MA_ = MA_.drop(MA_.tail(1).index)
# FLOAT --> INT
MA_.iloc[:,1:22] = MA_.iloc[:,1:22].astype(int)
# Save to CSV
MA_.to_csv(path_to)
```
